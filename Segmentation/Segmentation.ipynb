```python
import pandas as pd
from itertools import product
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
import scipy.integrate as scint
import itertools
import time
from itertools import product
from sklearn.metrics import mean_absolute_error
from scipy.stats import norm
import math
```

```python
def run_bee_eq_w1(x, ext_temp, truth, w, yy_param, error, b, treat, peri):
    def bee_eq(y, t, w, external_temperature, truth, theta_ideal, treat, peri):
        hive_temperature = y 
        u1 = [0]*len(hive_temperature)
        for i in range(len(hive_temperature)):
            if external_temperature[i] >= 0:
                u1[i] = 1
        if treat == 0:
            dydt = -w*hive_temperature*u1 + 2*(-hive_temperature + external_temperature)
        else:
            dydt = -w*hive_temperature*u1 + 2*(-hive_temperature) + external_temperature + peri
        return dydt
    
    t_max = len(ext_temp)-1
    t = np.linspace(0,t_max, num = t_max+1)
    params = []
    signals = []
    a = 10*np.ones(len(ext_temp))
    for i in range(len(w)):
        for j in range(len(yy_param)):
            truth_ = truth - yy_param[j]
            ext_temp_ = ext_temp - yy_param[j]
            peri_ = peri - yy_param[j]
            sol = scint.odeint(bee_eq, a, t, args=(w[i], ext_temp_, truth_, yy_param[j], treat, peri_)) # 'w' and 'amplitude_temp_ext'
            theta_t = sol[-1,:]
            error[b][j][i] = math.sqrt(np.mean((truth_ - theta_t)**2))
    return error
```

```python
def run_bee_eq_w2(x, ext_temp, truth, w, yy_param, error, b, treat, peri):
    def bee_eq(y, t, w, external_temperature, truth, theta_ideal, treat, peri):
        hive_temperature = y 
        u1 = [0]*len(hive_temperature)
        for i in range(len(hive_temperature)):
            if external_temperature[i] < 0:
                u1[i] = -1
        if treat == 0:
            dydt = w*hive_temperature*u1 + 2*(-hive_temperature + external_temperature)
        else:
            dydt = w*hive_temperature*u1 + 2*(-hive_temperature) + external_temperature + peri
        return dydt
    
    t_max = len(ext_temp)-1
    t = np.linspace(0,t_max, num = t_max+1)
    params = []
    signals = []
    a = 10*np.ones(len(ext_temp))
    for i in range(len(w)):
        for j in range(len(yy_param)):
            truth_ = truth - yy_param[j]
            ext_temp_ = ext_temp - yy_param[j]
            peri_ = peri - yy_param[j]
            sol = scint.odeint(bee_eq, a, t, args=(w[i], ext_temp_, truth_, yy_param[j], treat, peri_)) # 'w' and 'amplitude_temp_ext'
            theta_t = sol[-1,:]
            error[b][j][i] = math.sqrt(np.mean((truth_ - theta_t)**2))
    return error
```

```python
def run_bee_eq_w(x, ext_temp, truth, w, yy_param, treat, peri):
    def bee_eq(y, t, w, external_temperature, truth, theta_ideal, treat, peri):
        hive_temperature = y 
        u1 = [0]*len(hive_temperature)
        u2 = [0]*len(hive_temperature)
        for i in range(len(hive_temperature)):
            if external_temperature[i] >= 0:
                u1[i] = 1
            else:
                u2[i] = -1
        if treat == 0:
            dydt = -w[0]*hive_temperature*u1 + w[1]*hive_temperature*u2 + 2*(-hive_temperature + external_temperature)
        else:
            dydt = -w[0]*hive_temperature*u1 + w[1]*hive_temperature*u2 + 2*(-hive_temperature) + external_temperature + peri
        return dydt
    t_max = len(ext_temp)-1
    t = np.linspace(0,t_max, num = t_max+1)
    params = []
    signals = []
    a = 10*np.ones(len(ext_temp))
    truth_ = truth - yy_param
    ext_temp_ = ext_temp - yy_param
    peri_ = peri - yy_param
    sol = scint.odeint(bee_eq, a, t, args=(w, ext_temp_, truth_, yy_param, treat, peri_)) # 'w' and 'amplitude_temp_ext'
    theta_t = sol[-1,:]
    return theta_t
```

```python
def main(*argv):
    filename = argv[0]
    temp_data = argv[1]
    ext_data = argv[2]
    peri_data = argv[3]
    type_ = argv[4]
    data = pd.read_csv(filename)

    d = pd.DataFrame(data).groupby(data['Date'])
    w1 = np.linspace(0,50, num = 51)
    yy_param = np.linspace(min(data[temp_data]), max(data[temp_data]), 
                           num = np.int32(np.round((max(data[temp_data]) - min(data[temp_data]))/0.1)) + 1)
    error_w1 = [[[0 for i in range(len(w1))] for i in range(len(yy_param))] for i in range(len(d))]
    w2 = np.linspace(0,50, num = 51)
    error_w2 = [[[0 for i in range(len(w2))] for i in range(len(yy_param))] for i in range(len(d))]
    b = 0
    for i,j in d:
        x = np.linspace(0, len(j[temp_data].values)-1, num = len(j[temp_data]))
        j.reset_index()
        print(j['Date'].values[0])
        xx =j[ext_data].values
        yy = j[temp_data].values
        if type_ == 'Y':
            zz = j[peri_data].values
            treat = j['Treatment'].values[0]
        else:
            zz = j[temp_data].values
            treat = 0
        error_w1 = run_bee_eq_w1(x, xx , yy, w1, yy_param, error_w1, b, treat, zz)
        error_w2 = run_bee_eq_w2(x, xx , yy, w2, yy_param, error_w2, b, treat, zz)
        b = b+1

    w_s = list(itertools.product(w1,w2))

    error = [[[0 for i in range(len(w_s))] for i in range(len(yy_param))] for i in range(len(d))]

    for i in range(len(d)):
        print(i)
        for j in range(len(yy_param)):
            for m in range(len(w1)):
                for n in range(len(w2)):
                    error[i][j][n + m*51] = error_w1[i][j][m] + error_w2[i][j][n]

    pos = [0,len(d)]
    b = 1
    min_sig = []
    all_params = []
    all_positions = []
    all_lks = []
    while b < len(d):
        position = []
        params = []
        signals = []
        likelihood = []
        for a in range(len(d)):
            if a not in pos:
                positions = np.append(pos,a)
                positions.sort()
                sig_left = []
                left_params = []
                for i in range(len(positions)-1):
                    left_cut = np.unravel_index(np.array(error[positions[i]:positions[i+1]][:][:]).sum(axis = 0).argmin(), np.array(error[positions[i]:positions[i+1]][:][:]).sum(axis = 0).shape)
                    w_left = w_s[left_cut[1]]
                    yy_param_left = yy_param[left_cut[0]]
                    left_portion = data.iloc[positions[i]*24:positions[i+1]*24]
                    x_left_grp = left_portion.groupby(left_portion['Date'])
                
                    for j,k in x_left_grp:
                        x_left = np.linspace(0, len(k[temp_data].values)-1, num = len(k[temp_data]))
                        xx_left =k[ext_data].values
                        yy_left = k[temp_data].values
                        if type_ == 'Y':
                            zz_left = k[peri_data].values
                            treat = k['Treatment'].values[0]
                        else:
                            zz_left = k[temp_data].values
                            treat = 0
                        signal_left = run_bee_eq_w(x_left, xx_left , yy_left, w_left, yy_param_left, treat, zz_left)
                        sig_left.append(signal_left + yy_param_left)
                    left_params.append([w_left,yy_param_left])
                params.append(left_params)
                position.append(positions)
                signals.append(sig_left)
                L = np.sum(np.log(norm.pdf(data[temp_data].values.reshape(1,-1) - np.array(sig_left).reshape(1,-1))))
                likelihood.append(-2*L + 2*((len(positions)-2)+(len(positions)-1)*3))
        predicted = signals[np.argmin(likelihood)]
        all_params.append(params[np.argmin(likelihood)])
        all_positions.append(position[np.argmin(likelihood)])
        all_lks.append(likelihood[np.argmin(likelihood)])
        min_sig.append(predicted)
        pos = position[np.argmin(likelihood)]
        min_lks = likelihood[np.argmin(likelihood)]
        b = b+1

    lower = 0.1*min(all_lks)
    for i in range(len(all_lks)):
        if all_lks[i]<= min(all_lks) + lower:
            cut_points = all_positions[i]
            sig = min_sig[i]
            params = all_params[i]
            print(cut_points)
            print(params)
            for j in sig:
                for k in j:
                    print(k)
            break
```
